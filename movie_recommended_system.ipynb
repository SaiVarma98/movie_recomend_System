```python
import ast

def convert(obj):
    li=[]
    for i in ast.literal_eval(obj):
        li.append(i['name'])
    return li
```

```python
 movies['genres']=movies['genres'].apply(convert)
 movies['keywords']=movies['keywords'].apply(convert)
```

```python
# top 3 Actors from Cast

def convert3(obj):
    li=[]
    count=0
    for i in ast.literal_eval(obj):
        if count <=3:
            li.append(i['name'])
        else:
            break
        
    return li
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.iloc[0].crew
```

```python
def find_dire(obj):
    li=[]
    for i in ast.literal_eval(obj):
        if i["job"]=='Director':
            li.append(i['name'])
            break
    return li
        
```

```python
movies['crew']=movies['crew'].apply(find_dire)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.isnull().sum()
```

```python
# Removing Spaces between words of single person name to avoid confusion of tags
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x : [i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x : [i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
new_movies=movies.iloc[:,[0,1,7]]
```

```python
new_movies["tags"]=new_movies["tags"].apply(lambda x : " ".join(x))
```

```python
new_movies["tags"]=new_movies["tags"].apply(lambda x : x.lower())
```

```python
# Removing Stop words from the tags tab
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
import string
def message_text_process(sentence):
    no_punctuation=[char for char in sentence if char not in string.punctuation]
    no_punctuation=''.join(no_punctuation)
    return [word for word in no_punctuation.split() if word.lower() not in stopwords.words('english')]
```

```python
new_movies["tags"]=new_movies["tags"].apply(message_text_process).apply(lambda x : " ".join(x))
```

```python
new_movies.tail(3)
```

```python
# Applying Stemming 
from nltk.stem.porter import PorterStemmer
Ps=PorterStemmer()
```

```python
def stem(text):
    li=[]
    for i in text.split():
        li.append(Ps.stem(i))
    return " ".join(li)
```

```python
new_movies['tags']=new_movies['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer=CountVectorizer(max_features=5000)# can use stopwords='english' to implict removal of stop words
```

```python
vectors=vectorizer.fit_transform(new_movies['tags']).toarray() # default output scipy matrix so converting to array
```

```python
similarity=cosine_similarity(vectors)
```

```python
def recommend(movie):
    mv_index=new_movies[new_movies['title_x']==movie].index[0]
    dis=similarity[mv_index]
    mv_list=sorted(list(enumerate(dis)),reverse=True,key=lambda x:x[1])[1:6]
    for i in mv_list:
        print(new_movies.iloc[i[0]].title_x)
```

```python
def fetch_poster(id):
    response=rt.get("https://api.themoviedb.org/3/movie/{}?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US".format(id))
    data=response.json()
    return "https://image.tmdb.org/t/p/w500/"+data['poster_path']
```

```python
import pickle
import streamlit as st
import pandas as pd
import requests as rt
```

```python
st.title("Movie Recommended System")
movies_list=pickle.load(open("mv_list.pk1","rb"))
similarity=pickle.load(open("similarity.pk1","rb"))
mv_list=movies_list['title_x']
selected_name=st.selectbox('Enter any movie name', (mv_list))
```

```python
def recommend(movie):
    mv_index=movies_list[movies_list['title_x']==movie].index[0]
    dis=similarity[mv_index]
    mv_list=sorted(list(enumerate(dis)),reverse=True,key=lambda x:x[1])[1:6]
    recommended_movies=[]
    posters=[]
    for i in mv_list:
        movie_id=movies_list.iloc[i[0]].id
        posters.append(fetch_poster(movie_id))
        recommended_movies.append(movies_list.iloc[i[0]].title_x)
    return recommended_movies,posters
```

```python
if st.button('Recommend'):
    names,posters=recommend(selected_name)
    
    col1,col2,col3,col4,col5=st.columns(5)
    with col1:
        st.text(names[0])
        st.image(posters[0])   
    with col2:
         st.text(names[1])
         st.image(posters[1])   
    with col3:
          st.text(names[2])
          st.image(posters[2])   
    with col4:
           st.text(names[3])
           st.image(posters[3])   
    with col5:
            st.text(names[4])
            st.image(posters[4])  
```
